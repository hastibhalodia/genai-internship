```python
import torch
import matplotlib.pyplot as plt
from bigram_lm import BigramLanguageModel, vocab_size, decode

device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
# Load model
model = BigramLanguageModel(vocab_size)
model.load_state_dict(torch.load("bigram_model.pth", map_location=device))
model.to(device)
model.eval()
```

```python
# Plot training loss
loss_history = torch.load("loss_history.pth")
steps, train_loss, val_loss = zip(*loss_history)

plt.plot(steps, train_loss, label='Train')
plt.plot(steps, val_loss, label='Validation')
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate text
context = torch.zeros((1, 1), dtype=torch.long).to(device)
generated = model.generate(context, max_new_tokens=300)[0].tolist()
print(decode(generated))
```
